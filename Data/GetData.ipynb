Prepocess 20 Newsgroups data in the same way the 20 newsgroups datasest provided by octis is preprocessed

```python
from sklearn.datasets import fetch_20newsgroups
twng_sklearn = fetch_20newsgroups(subset = "all")["data"]
```

```python
with open("Raw_Data/twng_data.pickle", "wb") as file:
  pickle.dump(twng_sklearn, file)

twng_sklaern2 = [doc.replace('\n', " ").replace('\r', '') for doc in twng_sklearn]
twng_sklaern2 = [doc.replace('\t', " ") for doc in twng_sklaern2]
```

```python
with open("Raw_Data/twng_data.pickle", "wb") as file:
  pickle.dump(twng_sklaern2, file)
```

```python
with open('Auxillary_Data/twng_textData.txt', 'w') as handle:
    for line in twng_sklaern2:
        handle.write("%s\n" % line)
```

```python
preprocessor = Preprocessing(
    min_df = 0.005, 
    min_words_docs = 5
) 
```

```python
%%time
dataset_20ng = preprocessor.preprocess_dataset('Auxillary_Data/twng_textData.txt')
```

```python
dataset_20ng._Dataset__vocabulary = sorted(dataset_20ng._Dataset__vocabulary )
```

```python
import numpy as np
corpus_raw = np.array(twng_sklaern2)
corpus_octis = np.array(dataset_20ng.get_corpus())
```

```python
re_idx = np.array(dataset_20ng._Dataset__original_indexes)
```

### Reuters Data

Prepocess Reuters data with default settings in Octis

Raw Reuters corpus from https://archive.ics.uci.edu/ml/datasets/reuters-21578+text+categorization+collection

```python
# load Reuters textdata 
with open("Raw_Data/reuters_data.pickle", "rb") as f:
    dataset = pickle.load(f)

reuters_text = [' '.join(text.splitlines()) for text in dataset['data_text']]
```

```python
# Write reuters text data to file because the Octis interface for Preprocessing needs this representation
with open('Auxillary_Data/reuters_textData.txt', 'w') as handle:
    handle.write('\n'.join(reuters_text))
```

```python
preprocessor = Preprocessing(
  min_df = 0.005, 
  min_words_docs = 5
) # use Preprocessor by Octis with default arguments 
```

```python
%%time
dataset_reuters = preprocessor.preprocess_dataset('Auxillary_Data/reuters_textData.txt')
```

```python
dataset_reuters._Dataset__vocabulary = sorted(dataset_reuters._Dataset__vocabulary )
```

```python
# Write Octis dataset to pickle 
with open("Preprocessed_Data/octis_dataset_reuters.pickle", "wb") as f:
    pickle.dump(dataset_reuters, f)
```
